```
import numpy as np 
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
```

```
x=np.arange(1,15,1)
y=np.arange(1990,2010,1)
sin = np.sin(np.linspace(-4, 4, 100))
ax=plt.subplot(2,2,1)
ax.plot(sin[::10],range(10),'--o',c='k',lw=0.5)
#ax.set_axis_on()
ax2=plt.subplot(2,2,2)
ax2.plot(range(50,60),range(1990,2000))
ax=ax.add_child_axes(ax2)
```

```
iris=load_iris()
print("the classification targets "+str(iris['target_names']))
print("the classification features "+str(iris.feature_names))
print("the shape of our data is "+str(iris['data'].shape))
print("the first 5 rows of our data are\n "+str(iris['data'][0:5]))
print("the last 5 rows of our data are\n "+str(iris['data'][-6:-1]))
print("the types of our data are\n "+str(type(iris['data'])))
Xtrain,Xtest,ytrain,ytest=train_test_split(iris['data'],iris['target'],stratify=iris['target'],random_state=0)
print("the shape of our training data is "+str(Xtrain.shape))
print("the shape of our training labels is "+str(ytrain.shape))
print("the shape of our testing data is "+str(Xtest.shape))
print("the shape of our testing labels is "+str(ytest.shape))

```

```
data=Xtrain
bins=20
figsize=(15,15)
marker='o'
cmap=ListedColormap(['#0000aa', '#ff2020', '#50ff50'])

color=ytrain
size=60
alpha=0.8
n=Xtrain.shape[1]
features=[]
for i in range(n):
  features.append(data[:,i])

fig ,axs=plt.subplots(n,n,figsize=figsize)
#fig.set_label("Scatter matrix with matplotlib")

for row in range(n):
  for column in range(n):
    if row==column:
     axs[row,column].hist(features[row],bins=bins)
    else :
       axs[row,column].scatter(features[column],features[row],marker=marker,cmap=cmap,c=color,alpha=alpha,s=size)
#axs[0,1].scatter(features[0],features[1],marker=marker,c=ytrain,cmap='bwr')
#plt.axes("off")
   
```
